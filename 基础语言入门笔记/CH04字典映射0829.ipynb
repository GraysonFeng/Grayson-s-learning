```
# 字典的元素遍历
```

```
poi={'name':'shop','city':"shanghai",'information':{'address':"somewhere","num":66663333}}
```

```
for keys in poi:
  print(keys)
print('--------')
for values in poi:#默认是字典的key值
  print(values)
```

```
for keys in poi:
  print(keys)
print('--------')
for values in poi.values():#默认是字典的key值
  print(values)
print('--------')
for keys in poi:
  print(poi[keys])#通过引用
```

```
# 返回键值对
for (k,v) in poi.items():# items返回键值对，所有元素
  print('key为： %s,values为：%s'%(k,v))
```
